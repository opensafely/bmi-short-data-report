```python
import pandas as pd
from IPython.display import display

pd.set_option('display.max_rows', 500)
pd.options.mode.chained_assignment = None 
```

```python
### CONFIGURE ###
covariates = ['age_band','sex','ethnicity','region','imd','dementia','diabetes','hypertension','learning_disability']
count_columns = ['recorded_bmi_count',
       'weight_snomed_count', 'weight_backend_count', 'height_snomed_count',
       'height_backend_count']
aggregates = ['min','median','mean','max','std']
```

# Count of BMI-related events
## Count per patient

```python
df_patient_counts = pd.read_feather(f'../output/data/input_eventcounts.feather')
df_patient_counts[count_columns].describe()
```

```python
for cov in covariates:
    display(df_patient_counts.groupby(cov)[count_columns].agg(aggregates))
```
